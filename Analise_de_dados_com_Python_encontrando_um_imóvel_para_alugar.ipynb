```python
#Importanto bibliotecas necessárias
import pandas as pd
import seaborn as sns
```

```python
#importando arquivo csv do excel e dando nome a ele
aluguel_casa = pd.read_excel('/content/houses_rent.xlsx')

#olhando os 10 primeiros dados das colunas
aluguel_casa.head(10)
```

```python
#olhando os últimos dados
aluguel_casa.tail()
```

```python
#Renomeando nome de algumas colunas utilizando underline no lugar do espaço
aluguel_casa = aluguel_casa.rename(columns={'parking spaces':'parking_spaces', 'rent amount':'rent_amount','property tax':'property_tax','fire insurance':'fire_insurance'})

aluguel_casa.head(10)
```

```python
#Visualizando tipos de dados
aluguel_casa.dtypes
```

```python
#Modificando os dados de algumas colunas para o tipo Float
aluguel_casa = aluguel_casa.astype({'rent_amount':'float','property_tax':'float','fire_insurance':'float','total':'float'})

aluguel_casa.dtypes
```

```python
#Visualizando os dados modificados, perceba que os dados de algumas colunas agora estão no formato float
aluguel_casa.head(10)
```

```python
#Uma forma mais completa de obter informações dos dados, contendo contagem de cada categoria de dados, o quanto que
#consome de memória, número de linhas, quantidade de nulos e quantidade de colunas
aluguel_casa.info()
```

```python
#Acessando apenas as colunas com números para obter dados estatísticos
aluguel_casa[['area','rooms','bathroom','parking_spaces','rent_amount','property_tax','fire_insurance','total']].describe()
```

```python
#Acessando os 5 primeiros valores de cada coluna com o valor do aluguel igual ou abaixo de 4000
aluguel_casa = aluguel_casa[(aluguel_casa['total']<=4000)]

aluguel_casa.head()
```

```python
#Adicionar os imóveis que aceitam animais e com um valor médio de 4000 ou menos de aluguel e mostrar os 10 primeiros resultados
aluguel_casa = aluguel_casa[(aluguel_casa['total']<=4000) & (aluguel_casa['animal']=='accept')]

aluguel_casa.head()
```

```python
#Na mesma base do codigo anterior vamos colocar mais um filtro, será o de numeros de quarto entre 2 e 3 quartos
#ficando os filtros de: aluguel menos de 4000, que aceite animais e que tenha 2 e 3 quartos

aluguel_casa = aluguel_casa[(aluguel_casa['total']<=4000) & (aluguel_casa['animal']=='accept') & ((aluguel_casa['rooms']==2)|(aluguel_casa['rooms']==3))]
aluguel_casa.head()


```

```python
aluguel_casa.describe()
```

```python
#-------------------------------------------------------------------------------------------------------------------------------------#
```

```python
#"Filtros mais simplificados"
```

```python
#Para agilizar o processo de tratamento de dados temos outras maneiras de obter o memso resultado, porém de forma mais simplificada
#Vamos refazer algumas estapas utilizando outras metodologias de códigos:
```

```python
#Nesse exemplo vamos filtrar os preços de aluguel menor que 4000 e se aceita animal, perceba que o codigo é mais simplificado utilizando a função ""query"""
aluguel_casa = aluguel_casa.query("total <= 4000 and animal == 'accept'")  #obs: ficar atento com as aspas, as de fora sempre duplas e as internas uma
aluguel_casa.head()
```

```python
#nesse outro exemplo adicionamos o numero de quartos para filtrar também com 2 e 3 quartos com a função ""query""
aluguel_casa = aluguel_casa.query(("total <= 4000 and animal =='aceept'" and 'rooms == 2 or rooms == 3'))
aluguel_casa.head()
```

```python
#----------------------------------------------------------------------------------------------------------------------------------------------#
```

```python
 #"Agregando dados"
```

```python
#Sabemos a importância de agegar dados, com isso conseguimos mais informações e extraimos mais insights
#iremos aprender esse processo no python para ganhar uma gama maior de informações
#Vamos agregar os dados de city para tirar a media total dos valores com a função ""groupby""" para juntar os dados de city e total, depois "".agg('mean)"" para tirar a media
#finalizando com: "".reset_index()"" para deixar em formato DF

agg_city_price = aluguel_casa.groupby(['city'])['total'].agg('mean').reset_index()
agg_city_price
```

```python
#Agrupando dois grupos que seria bathrooms e room para tirarmos a média de todas as variações deles tipo: 1 banheiro e 2 quartos ou 2 e 4 banheiros e assim por diante.
agg_bath_room_price = aluguel_casa.groupby(['bathroom', 'rooms'])['total'].agg('mean').reset_index()
agg_bath_room_price
```

```python
#----------------------------------------------------------------------------------------------------------------------------------------#
```

```python
#Iniciaremos a visualização dos dados, importaremos a biblioteca ""seabor"" para vuzulizar graficos, isso facilitará as nossas análises, nosso cérebro tem
#a capacidade de atravez de imagens observar melhor os insights

import seaborn as sns
```

```python
#Alguns parâmetros vamos introduzir para executar o gráfico: "" ax"" ele ajuda a ordenar o que vc quer fazer nos eixos mudar nome etc...
#""sns.barplot"" está pedindo para inserir um grafico, depois selecionei o que eu quero montar o gráfico, em seguida os eixos que ficarão os dados
#""color =""para pintar da cor desejada e por último renomear com a função ax.set e dentro da função modifico o nome das colunas, caso queira tirar os dados superiores
#do gráfico utilizar ponto e virgula ; depois do código
ax = sns.barplot(data = agg_bath_room_price, x = 'rooms', y= 'total', color = 'blue');
ax.set(xlabel = 'Quantidadae de Quartos', ylabel = 'Média do Preço');
```

```python
#Utilizando a mesma base de código vamos analisar a quantidade de banheiros com o valor na média dos 4000
ax = sns.barplot(data = agg_bath_room_price, x = 'bathroom', y= 'total', color = 'green');
ax.set(xlabel = 'Quantidadae de Banheiros', ylabel = 'Média do Preço');
```

```python
#Analisando uma situação: referente ao desvio padrão sobre área com valores de até 4000, percebemos que existia uma dispersão muito grande sobre a média
#vamos plotar um grafico representando isso visualmente:
agg_area = aluguel_casa[['area']] #Aqui selecionamos apenas a área
sns.histplot(data=agg_area, x='area');
```

```python
#nesse grafico percebemos que as movimentações estão dentro de apenas 2000 para baixo
#vamos aplicar o ""describe""para entender melhor os dados:
agg_area.describe() #como mencionado, desvio padrão sobre área está muito distante da média, para melhorar a visualização
                    #vamso colocar um outro filtro com a média de área abaixo de 100 metros, afinal corresponde 75% dos dados


```

```python
#Filtrando a base de dados de area para igual ou menor que 100 metros de area
agg_area = agg_area.query('area <= 100')
agg_area.describe()   #podemos observar que agora o desvio padrão ficou menor que a média
                      #isso acontece porque agora não temos mais os dados de áreas maiores, fizemos essa base dentro dos 75%
                      #agora temos uma média e maximo e 75% com os valores próximos
```

```python
#Vamos plotar um novo gráfico e ver como ele vai ser representado depois dessas mudanças:
sns.histplot(data=agg_area, x='area');   #Perceba que depois de filtrarmos os dados para 100 metros, conseguimos ver os dados de uma forma mais clara
                                         #podemos observar agora que temos um pico onde 55-60 metros as pessoas alugam mais
```

```python
#agora de uma maneira mais ampla, vamos analizar a média subdividido pelas cidades de Barcelona e Porto ficando dessa forma:
sns.histplot(data = aluguel_casa, x='total', hue ='city'); #observe essa nova função ""hue"" serve para colocar subcategorias, fazendo um grafico representado pelas duas cidades com duas cores
```

```python
#------------------------------------------------------------------------------------------------------------------------------------------------#
```

```python
#Agora que já temos os dados de metragem mais procuradas e valores de alugueis da cidade
#agora vamos escolher o apartamento para alugar melhor custo benefício, vamos selecionar a cidade porto
aluguel_casa = aluguel_casa.query("city =='Porto'")
aluguel_casa.head()
```

```python
aluguel_casa.describe()
```

```python
#Agora vamos organizar a quantidade de apartamentos e em quais andares tem as quantidades
#Vamos agregar os dados de floor para contar uma unica vez cada andar a venda com:""index.nunique()"" depois "".sort_values(ascending = False) para aparecer os valores de forma decrescente
aluguel_casa.groupby('floor').index.nunique().sort_values(ascending = False)
```

```python
#Agora que sabemos a quantidade de apartamentos por andares plotado no códico anterior, agora vamos tirar a média dos valores por andar
agg_floor_price = aluguel_casa.groupby('floor')['total'].agg('mean').reset_index()
agg_floor_price
```

```python
#Agora que temos a média de valores por andares diponíveis, vamos visualizar isso de uma forma de gráfico de barras para interpretar melhor...
sns.barplot(data = agg_floor_price, x='floor', y='total', color = 'blue');

#Perceba que depois de plotado o gráfico podemos visualizar alguns dados interessante sobre conseguir morar em um andar alto pagando o mesmo preço ou
#até mesmo pagando menos que os andares inferiores, no caso o andar 16 está sendo bem vantajoso perando aos de mais com a mesma media de preço
```

```python
#Agora com base dos dados anteriores vamos ver as opções que temos dos apartamentos que ficam no andar 16 e qual será mais vantajoso:
sixtheen_floor = aluguel_casa.query('floor==16')
sixtheen_floor.head()
```
